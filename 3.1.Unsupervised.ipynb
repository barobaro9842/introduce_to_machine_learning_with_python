## 3.3.1. 전처리 방법

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

cancer = load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=1)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler # 미리 구현된 전처리기

scaler = MinMaxScaler()

scaler.fit(X_train) #비지도 학습
```

학습 결과를 바탕으로 실제로 스케일 변환을 실행하려면 transform 함수를 사용해야함

```python
X_train_scaled = scaler.transform(X_train)
print("shape after transform : ", X_train_scaled.shape)
print("min before scaling : \n", X_train.min(axis=0))
print("max before scaling : \n", X_train.max(axis=0))
print("min after scaling : \n", X_train_scaled.min(axis=0))
print("max after scaling : \n", X_train_scaled.max(axis=0))
```

다양했던 특성별 최솟값과 최댓값이 전부 0과 1로 통일

```python
# 테스트 세트 변환

X_test_scaled = scaler.transform(X_test)

print("max after scaling : \n", X_test_scaled.min(axis=0))
print("min after scaling : \n", X_test_scaled.max(axis=0))
```

train 데이터를 기준으로 min과 max값이 결정되었기 때문에 test셋에서는 0과 1이 아닌 다른 결과가 도출된다.   
스케일링을 할 때는 반드시 하나의 기준(train)으로 모든 데이터를 해야한다.

## 3.3.3. Quantile Transformer & PowerTransformer 

```python
from sklearn.datasets import make_blobs
from sklearn.preprocessing import QuantileTransformer, StandardScaler, PowerTransformer
```

```python
X, y = make_blobs(n_samples=50, centers=2, random_state=4, cluster_std=1)
X += 3

plt.scatter(X[:, 0], X[:, 1], c=y, s=30, edgecolors='black')
plt.xlim(0, 16)
plt.xlabel('x0')
plt.ylim(0, 10)
plt.ylabel('x1')
plt.title("Original Data")
plt.show()
```

#### Quantile Transformer 적용

```python
scaler = QuantileTransformer(n_quantiles=50)
X_trans = scaler.fit_transform(X)

plt.scatter(X_trans[:, 0], X_trans[:, 1], c=y, s=30, edgecolors='black')
plt.xlim(0, 1.25)
plt.xlabel('x0')
plt.ylim(0, 1.25)
plt.ylabel('x1')
plt.title("Original Data")
plt.show()
```

```python
fig, axes = plt.subplots(1,2,figsize=(12,4))
axes[0].hist(X)
axes[0].set_title('X')
axes[1].hist(X_trans)
axes[1].set_title('X_trans')
plt.show()
```

```python
scaler.quantiles_
```

#### Quantile Transformer 동작의 이해

```python
x = np.array([[0], [5], [8], [9], [10]])
print(np.percentile(x[:, 0], [0, 25, 50, 75, 100]))
```

입력된 변수의 길이와 분위수가 같기 때문에 입력 변수와 같은 값이 도출된다.

```python
scaler = QuantileTransformer(n_quantiles=5).fit(x)
x_trans = scaler.transform(x)
print(scaler.quantiles_)
print(np.percentile(x_trans[:,0], [0, 25, 50, 75, 100]))
```

모든 데이터가 균등한 간격으로 재배치 된것을 알 수 있다.

## 3.3.4. 훈련데이터와 테스트 데이터의 스케일을 같은 방법으로 조정하기

```python
from sklearn.datasets import make_blobs
from sklearn.preprocessing import MinMaxScaler

#인위적인 데이터셋 생성
X, _ = make_blobs(n_samples=50, centers=5, random_state=4, cluster_std=2)
X_train, X_test = train_test_split(X, random_state=5, test_size=.1)

#scatter plot
fig, axes = plt.subplots(1, 3, figsize=(13,4))
axes[0].scatter(X_train[:,0], X_train[:,1], c=mglearn.cm2.colors[0], label="training set", s=60)
axes[0].scatter(X_test[:,0], X_test[:,1], c=mglearn.cm2.colors[1], label="test set", s=60)
axes[0].legend(loc='upper left')
axes[0].set_title('original data')

#Minmax Scaler
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

#scaled scatter plot
axes[1].scatter(X_train_scaled[:,0], X_train_scaled[:, 1], c=mglearn.cm2.colors[0], label="training set", s=60)
axes[1].scatter(X_test_scaled[:,0], X_test_scaled[:, 1], c=mglearn.cm2.colors[1], label="test set", s=60)
axes[1].set_title('scaled data')

#wrong scaling(train, test separately)
train_scaler = MinMaxScaler()
test_scaler = MinMaxScaler()
X_train_mal_scaled = train_scaler.fit_transform(X_train)
X_test_mal_scaled = test_scaler.fit_transform(X_test)

axes[2].scatter(X_train_mal_scaled[:,0], X_train_mal_scaled[:,1], c=mglearn.cm2.colors[0], label='training set', s=60)
axes[2].scatter(X_test_mal_scaled[:,0], X_test_mal_scaled[:,1], c=mglearn.cm2.colors[1], label='test set', s=60)
axes[2].set_title('wrong scaling')

for ax in axes :
    ax.set_xlabel('feat 0')
    ax.set_ylabel('feat 1')
```

## 3.3.5. 지도 학습에서 데이터 전처리 효과

```python
from sklearn.svm import SVC
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=0)

svm = SVC(gamma='auto')
svm.fit(X_train, y_train)
print("test set acc : {:.2f}".format(svm.score(X_test, y_test)))
```

#### scaling

```python
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

svm.fit(X_train_scaled, y_train)

print("test set acc (scaled) : {:.2f}".format(svm.score(X_test_scaled, y_test)))
```

```python
# standardscaler = avg : 0, var : 1
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

#SVM 학습
svm.fit(X_train_scaled, y_train)

print("SVM test accuracy : {:.2f}".format(svm.score(X_test_scaled, y_test)))
```
